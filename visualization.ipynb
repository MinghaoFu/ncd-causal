```python


import matplotlib.pyplot as plt
import numpy as np
import os

vis_dir = "vis_dir"
os.makedirs(vis_dir, exist_ok=True)  # Create the directory if it doesn't exist


# Data
alpha_labels = [512, 768, 1024, 1536, 2048, 3072, 4096, 6144]
all_acc = [37.8, 38.8, 38.4, 39.2, 38.4, 39.7, 36.2, 34.6]

# Convert labels to equidistant positions
x_positions = np.arange(len(alpha_labels))  # Generate equidistant positions

# Create the bar plot
plt.bar(x_positions, all_acc, color='teal', edgecolor='black', width=0.6)

# Set custom x-ticks
plt.xticks(x_positions, alpha_labels, fontsize=14)

# # Set labels and title
# plt.xlabel(r'$n_{zs}$', fontsize=16)
# plt.ylabel('All ACC (%)', fontsize=16, labelpad=10)
# plt.title(r'$n_{zs}$ - CUB', fontsize=18)

# Set Y-axis limits and ticks
plt.ylim(34, 40)
plt.yticks([34, 35, 36, 37, 38, 39, 40], fontsize=14)

# Adjust the spines for a similar appearance
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_linewidth(1.5)
plt.gca().spines['bottom'].set_linewidth(1.5)

# Customize tick sizes
plt.tick_params(axis='both', which='major', labelsize=14, width=1.5)

# Add grid for minor aesthetics (optional)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout for cleaner presentation
plt.tight_layout()

save_path = os.path.join(vis_dir, "n_zs.png")
plt.savefig(save_path, dpi=300, bbox_inches='tight')  # High resolution, tight layout
```
